```python
df2 = pd.read_csv('/content/drive/MyDrive/capstone_yb/Voting ML/DATA/overlapping_matching2.csv')
INCOME= pd.read_csv('/content/drive/MyDrive/capstone_yb/Voting ML/DATA/filtered_income.csv')
```

```python
INCOME
```

```python
# df1과 df2의 'Geographic Area Name' 열의 고유한 값 출력
print("Unique values in df1 'Geographic Area Name' column:")
print(INCOME['Geographic Area Name'].unique())

print("\nUnique values in df2 'Geographic Area Name' column:")
print(df2['Geographic Area Name'].unique())
```

```python
# df2의 'Geographic Area Name' 열의 각 값에서 뛰어쓰기 제거
INCOME['Geographic Area Name'] = INCOME['Geographic Area Name'].str.replace(' ', '')
df2['Geographic Area Name'] = df2['Geographic Area Name'].str.replace(' ', '')
# County 열을 할당

# 두 DataFrame을 'Geographic Area Name' 열을 기준으로 병합하기
merged_INB = pd.merge(INCOME, df2, on='Geographic Area Name')
merged_INB.rename(columns={'Unnamed: 0': 'Vote Count'}, inplace=True)

# 새로운 CSV 파일로 저장
merged_INB.to_csv('/content/drive/MyDrive/capstone_yb/Voting ML/DATA/merged_INB.csv', index=False)

print("Merged file saved successfully.")
```

```python
# Save Final File
df3 = pd.read_csv('/content/drive/MyDrive/capstone_yb/Voting ML/DATA/overlapping_matching_avg_biden.csv')

# df2의 'Geographic Area Name' 열의 각 값에서 뛰어쓰기 제거
INCOME['Geographic Area Name'] = INCOME['Geographic Area Name'].str.replace(' ', '')
df3['Geographic Area Name'] = df3['Geographic Area Name'].str.replace(' ', '')
# County 열을 할당

# 두 DataFrame을 'Geographic Area Name' 열을 기준으로 병합하기
merged_INAV = pd.merge(INCOME, df3, on='Geographic Area Name')
merged_INAV.rename(columns={'Unnamed: 0': 'Vote Count'}, inplace=True)

# 새로운 CSV 파일로 저장
merged_INAV.to_csv('/content/drive/MyDrive/capstone_yb/Voting ML/merged_INAV.csv', index=False)

print("Merged file saved successfully.")
```

```python
# 데이터프레임에서 숫자형으로 변환해야 할 열들 선택
numeric_columns = [
    'House Hold Less than $50,000',
    'House Hold $50,000 to $99,999',
    'House Hold $100,000 to $149,999',
    'House Hold $150,000 to $199,999',
    'House Hold $200,000 or more',
    'Families Less than $50,000',
    'Families $50,000 to $99,999',
    'Families $100,000 to 149,999',
    'Families $150,000 to $199,999',
    'Families $200,000 or more',
    'Married-couple Families Less than $50,000',
    'Married-couple Families $50,000 to $99,999',
    'Married-couple Families $100,000 to 149,999',
    'Married-couple Families $150,000 to $199,999',
    'Married-couple Families $200,000 or more',
    'Nonfamily households Less than $50,000',
    'Nonfamily households $50,000 to $99,999',
    'Nonfamily households $100,000 to 149,999',
    'Nonfamily households $150,000 to $199,999',
    'Nonfamily households $200,000 or more'
]

# 숫자형으로 변환
for column in numeric_columns:
    merged_INAV[column] = pd.to_numeric(merged_INAV[column], errors='coerce')

# 누락된 값(NaN)을 0으로 대체
merged_INAV.fillna(0, inplace=True)

# Most_Preferred_Group 열 추가
merged_INAV['Most_Preferred_Group'] = merged_INAV[numeric_columns].idxmax(axis=1)

# 결과 확인
result_df = merged_INAV[['Geographic Area Name', 'Most_Preferred_Group', 'average_Biden_proportion']]
print(result_df)
```
