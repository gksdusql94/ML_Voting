```python
df = pd.read_csv('/content/drive/MyDrive/capstone_yb/Voting ML/DATA/WA_Income.csv')
# Margin of Error
df.drop(columns=df.filter(like='Margin of Error').columns, inplace=True)
df.shape
```

```python
import numpy as np
# (X) >> NaN Change
df.replace('(X)', np.nan, inplace=True)
# ('N') >> NaN Change
df.replace('N', np.nan, inplace=True)
missing_values = df.isnull().sum() / len(df)
print(missing_values)

# Missing Ration is not 0
missing_columns = missing_values[missing_values > 0]

print("Missing Ratio:")
print(missing_columns)
```

```python
# Delete missing columns
# Drop columns with missing ratios of 1.0
df.drop(columns=missing_columns.index, inplace=True)
df.head()
```

```python
# 'PERCENT ALLOCATED'이 포함된 열 찾기
cols_to_drop = df.columns[df.columns.str.contains('PERCENT ALLOCATED')]

# 'PERCENT ALLOCATED'이 포함된 열 삭제
df.drop(columns=cols_to_drop, inplace=True)

# 추가적으로 목록에 있는 열들 삭제
additional_cols_to_drop = [
    'Estimate!!Nonfamily households!!Total',
    'Estimate!!Married-couple families!!Total',
    'Estimate!!Families!!Total',
    'Estimate!!Households!!Total'
]

# 추가적인 열 삭제
df.drop(columns=additional_cols_to_drop, inplace=True)
```

```python
# Save the cleaned DataFrame to a new CSV file
df.to_csv('cleaned_income.csv', index=False)
```
